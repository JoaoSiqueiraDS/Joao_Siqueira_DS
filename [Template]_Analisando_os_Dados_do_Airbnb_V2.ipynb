### **Variables correlation**
**Correlation** or **dependence** is a statistical relationship between two variables. Our objective here is to find it in our dataset.

This relationship can be measured, and the purpose of the correlation coefficient is to establish how strong is this relationship. There are only two procedures to identify the existent correlations in these variables:
* Generate a correlation matrix.
* Create a heatmap from this matrix by using `seaborn`library.

```
# Creating a correlation matrix
correlation = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365' ]].corr()

# Displaying the correlation matrix
display(correlation)
```

###Preparing the data

Let's take a look on this new dataset. First, let's import it. 

```
san_fran = pd.read_csv('http://data.insideairbnb.com/united-states/ca/san-francisco/2020-06-08/data/listings.csv.gz', compression='gzip')
san_fran.shape
```

```
san_fran.head(4)
```

We can see that the id column in this new dataframe matches this column in the df_clean dataframe. 

In this new dataframe, we are going to use only two columns: id and zipcode. Let's generate a dataset "zipcode" with these two columns. 

```
zipcode = san_fran[["id", "zipcode"]]
zipcode.shape
```

Let's now merge it into the `df_clean` dataset.

```
df_zip = df_clean.merge(right=zipcode, how="left", on="id", )
df_zip.info()
```

We have now a total of 7379 rows / entries. 
However, there are some null values for zip code. Before moving on, it is important to check if the number of null values is considerable. Let's do it.

```
df_zip['zipcode'].isnull().sum()/len(df_zip)*100
```

There aren't many null values and so that we are going to drop them. 

```
df_zip.dropna(subset=["zipcode"], axis = 0, inplace=True)
print("The number o null values are {}".format(df_zip["zipcode"].isnull().sum()))
df_zip.shape
```

Ok, We got rid of the null values and there are still 5696 rows left in our dataframe. Let's check the "zipcode" column to see if it needs cleaning. 

```
df_zip["zipcode"].unique()
```

We can see that some zipcodes present a "CA" before the number. It needs removing. There is also one value that presents only "CA" and of course it will also be removed. 

```
#Eliminating the element "CA".
#First thing is to get the index for which column zipcode has value "CA".
indexes = df_zip[df_zip["zipcode"]=="CA"].index
#Deleting the rows indexes from dataframe
df_zip.drop(indexes, inplace=True)

df_zip["zipcode"].unique()

```

```
#Converting the whole column to string and then removing the "CA".
df_zip["zipcode"] = df_zip["zipcode"].str.replace('CA','').str.strip()

df_zip['zipcode'].unique()

```

```
#Converting the dataset into string
df_zip["zipcode"] = df_zip["zipcode"].astype(str)
df_zip["zipcode"].unique()
```

```
#Checking if there is empty values
df_zip[df_zip["zipcode"]=="nan"]
```

```
#droping the rows featuring zipcode "nan"
df_zip = df_zip[df_zip["zipcode"]!="nan"]
df_zip.reset_index(drop=True, inplace = True)
df_zip.shape
```

Great! Now we have our dataset absolutely clean. Let's plot the map with **Folium**

To do so, we are going to do the following steps:

* Create a figure
* Create a map
* Create the [Choropleth]("https://en.wikipedia.org/wiki/Choropleth_map") map.
* Create some markers on the map pointing to some famous attractions in San Francisco.
* Display the map


```
#Create a figure
figure = fl.Figure(width=750, height=750)

#Create the map
map = fl.Map(location=[37.773972, -122.431297], min_zoom=10)

#Crate the Choropleth
SF_Choropleth = fl.Choropleth(
    geo_data = "https://data.sfgov.org/api/geospatial/u5j3-svi6?method=export&format=GeoJSON",
    data = df_zip,
    columns = ["zipcode","price"],
    key_on = "feature.properties.zip",
    fill_color ='YlOrRd',
    highlight = True,
    bins = 6,
    nan_fill_color = "white",
    fill_opacity = 0.7,
    line_opacity = 0.3,
    legend_name = "Mean Price",
    show=True

).add_to(map)

#Diplay the map
figure.add_child(map)
map
```

```
#Defining a function to plot the map
def map_feature_by_zipcode(df_zip, col):
    """
    Generates a folium map of Seattle
    :param zipcode_data: zipcode dataset
    :param col: feature to display
    :return: m
    """

    # read updated geo data
    king_geo = "cleaned_geodata.json"

    # Initialize Folium Map with Seattle latitude and longitude
    m = fl.Map(location=[37.773972, -122.431297], min_zoom=10,
                   detect_retina=True, control_scale=False)

    # Create choropleth map
    m.choropleth(
        geo_data=king_geo,
        name='choropleth',
        data=df_zip,
        # col: feature of interest
        columns=['zipcode', col],
        key_on='feature.properties.ZIPCODE',
        fill_color='OrRd',
        fill_opacity=0.9,
        line_opacity=0.2,
        legend_name='Mean ' + col
    )

    fl.LayerControl().add_to(m)

    # Save map based on feature of interest
    m.save(col + '.html')

    return m
```
